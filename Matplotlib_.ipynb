Scatter plot with colormap

```python
x=np.array([1,2,6,8,3,6,4,7,5,2,6])
y=np.array([10,50,84,63,74,21,53,87,10,65,48])
colors=np.array(['red', 'green', 'blue', 'yellow', 'orange', 'purple', 'pink', 'brown', 'black', 'white','aqua'])
size=np.array([100,200,500,400,300,150,250,120,150,300,100])

plt.scatter(x,y,c=colors,s=size)
plt.show()
```

```python
data=pd.DataFrame({
    'x':np.random.randn(100),
    'y':np.random.randn(100),
    'value':np.random.randn(100)
})

cmap="viridis"
alpha=1
plt.figure(figsize=(6,6))
plt.title=("scaatter plot with colormap")
plt.scatter(data['x'],data['y'],c=data['value'],cmap=cmap,alpha=alpha)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.colorbar(label='value')
plt.show()
```

# contour plot

```python
data=pd.DataFrame(np.random.rand(10,10))
plt.figure(figsize=(10,16))
plt.contour(data.values)
plt.show()
```

#Bagraph

1.   it is used to display the category of data with bars with length and height set by user


```python
section=['A','B','C','D']
Student=[10,20,30,40]
plt.bar(section,Student)
plt.show()
```

```python
#pichart
x=[10,20,30,40,60]
l=['apples','bannans','cherries','oranages','mangos']
plt.pie(x,labels=l)
plt.show()
```

```python
#pichart with Explode
Ex=[0.0,0.0,0.0,0.0,0.2]
plt.pie(x,labels=l,explode=Ex)
plt.show()
```

```python
#histogram
x=[1,2,2,3,4,4,5,6,6]
plt.hist(x)
plt.show()
```

```python
#subplot
section=['a','b','c']
student=[10,20,30]
marks=[10,20,20,30,40,40]
plt.subplot(2,2,1)
plt.bar(section,student)
plt.subplot(2,2,2)
plt.pie(student)
plt.subplot(2,2,3)
plt.hist(marks)
plt.subplot(2,2,4)
plt.plot([1,2,3],[4,5,6])
plt.show()
```

```python
#heatmap
data=np.random.random((13,16))
plt.imshow(data)
plt.title("2-D HeatMap")
plt.colobar()
plt.show()
```
